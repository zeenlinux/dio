```python
# é o pandas né pai
# permite montar um esquema de validação dos dataframes
import pandas as pd
import pandera as pa
```

## Carregamento do csv

```python
td = pd.read_csv("ocorrencia.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True) 
#no colchetes da para botar uma lista de colunas tipo data
#dayfirst=True ... organiza a tabela pelos dias
td.head(10) # primeiras 10 linhas
# td.tail(10) #mostra as ultimas
```

# Validação de dados

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo":pa.Column(pa.Int, required=False),
        "codigo_ocorrencia":pa.Column(pa.Int),
        "codigo_ocorrencia2":pa.Column(pa.Int),
        "ocorrencia_classificacao":pa.Column(pa.String),
        "ocorrencia_cidade":pa.Column(pa.String),
        "ocorrencia_uf":pa.Column(pa.String, pa.Check.str_length(2,2),nullable=True),
        "ocorrencia_aerodromo":pa.Column(pa.String),
        "ocorrencia_dia":pa.Column(pa.DateTime),
        "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes":pa.Column(pa.Int)
        
    }

)
#nullable=True representa q na tabela podem ter celulas nulas
#pa.Check.str_matches tem q ser o segundo
#str_length(2,2) tamanho min e max do dado de texto
#required=False caso eu queira reaproveitar o esquema porem das colunas q ja tem feito alguma nao estar no banco de dados
```

```python
schema.validate(td)
```
